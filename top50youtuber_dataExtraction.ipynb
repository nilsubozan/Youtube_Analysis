```python
#Youtube Data API Connection
api_key= 'YOUR_API_KEY'
api_service_name = "youtube"
api_version = "v3"  
youtube = build(api_service_name, api_version, developerKey=api_key)



def get_channel_stats(youtube,channel_ids):
    all_data= []

    #making API request
    request = youtube.channels().list(
        part="snippet,contentDetails,statistics",
        id=','.join(channel_ids)
    )
    response = request.execute()

    #processing API request into dataframe
    for item in response['items']:
        data={'channelName': item['snippet']['title'],
              'subscribers': item['statistics']['subscriberCount'],
              'views': item['statistics']['viewCount'],
              'totalVideos': item['statistics']['videoCount'],
              'playlistId': item['contentDetails']['relatedPlaylists']['uploads']
        }
        
        all_data.append(data)
        
    return(pd.DataFrame(all_data))
        

```

```python
channel_stats= get_channel_stats(youtube,channel_ids)
```

```python
channel_stats['subscribers'] = pd.to_numeric(channel_stats['subscribers'], errors='coerce')
```

```python
#Sorting the channels in descending order considering number of subscribers
channel_stats2=channel_stats.sort_values(by='subscribers',ascending=False)
```

```python
import time
import json
from googleapiclient.discovery import build

#Getting video IDs from playlists with rate limiting and incremental saving
def get_video_ids_from_playlists(youtube, playlist_ids, save_interval=3, output_file='video_ids.json'):  # Changed save_interval to 5
    playlist_video_data = {}  #For each playlist(youtube channel), video ids will be stored in dictionary

    #Tring to loading the previously saved data to avoid duplicating work
    try:
        with open(output_file, 'r') as f:
            playlist_video_data = json.load(f)
            print(f"Loaded saved data from {output_file}")
    except FileNotFoundError:
        print("No existing data found, starting from scratch.")

    #Extracting video_ids for each playlistID 
    for idx, playlist_id in enumerate(playlist_ids):
        print(f"Processing playlist {idx+1}/{len(playlist_ids)}: {playlist_id}")
        if playlist_id in playlist_video_data:
            print(f"Playlist {playlist_id} already processed, skipping...")
            continue

        #Youtube Data API only allows 50 video_ids, therefore pagination is implemented to handle more than 50 videos
        video_ids = []
        try:
            request = youtube.playlistItems().list(
                part="snippet,contentDetails",
                playlistId=playlist_id,
                maxResults=50  
            )
            response = request.execute()
            
            #Added print statements to follow the process
            print(f"Received response for playlist {playlist_id}")

            #Extractings video IDs from the first page
            for item in response.get('items', []):
                video_ids.append(item['contentDetails']['videoId'])

            #Handle pagination in case there are more than 50 videos
            next_page_token = response.get('nextPageToken')

            while next_page_token is not None:
                request = youtube.playlistItems().list(
                    part="snippet,contentDetails",
                    playlistId=playlist_id,
                    maxResults=50,
                    pageToken=next_page_token
                )
                response = request.execute()

                #Extract video IDs from subsequent pages
                for item in response.get('items', []):
                    video_ids.append(item['contentDetails']['videoId'])

                #Added sleep() to prevent from Youtube Data API quota problems
                next_page_token = response.get('nextPageToken')
                time.sleep(2)  

        except Exception as e:
            print(f"Error while processing playlist {playlist_id}: {e}")
            continue

        playlist_video_data[playlist_id] = video_ids

        #Save progress incrementally
        if (idx + 1) % save_interval == 0:
            print(f"Processed {idx + 1} playlists, saving progress...")
            with open(output_file, 'w') as f:
                json.dump(playlist_video_data, f)
       
        time.sleep(2)  

    #Store final results in a json file
    with open(output_file, 'w') as f:
        json.dump(playlist_video_data, f)

    return playlist_video_data


playlist_ids = top3channels['playlistId'].tolist() 

#Function call
playlist_video_data = get_video_ids_from_playlists(youtube, playlist_ids)

for playlist_id, video_ids in playlist_video_data.items():
    print(f"Playlist ID: {playlist_id} | Number of Videos: {len(video_ids)}")

```

```python
import json
import pandas as pd
from googleapiclient.discovery import build

#Loading the previously saved json files to extract video details
def load_video_ids_from_json(json_file):
    with open(json_file, 'r') as f:
        video_data = json.load(f)
    return video_data


def get_video_details(youtube, video_ids):
    all_video_info = []

    for i in range(0, len(video_ids), 50):
        #Will be loaded as a chunks of 50 for each API call
        request = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            id=','.join(video_ids[i:i+50])
        )
        response = request.execute()

        #Looping through the response
        for video in response['items']:
            stats_to_keep = {
                'snippet': ['channelTitle', 'title', 'description', 'tags', 'publishedAt'],
                'statistics': ['viewCount', 'likeCount', 'favoriteCount', 'commentCount'],
                'contentDetails': ['duration', 'definition', 'caption']
            }

            video_info = {}
            video_info['video_id'] = video['id']

            #Collect only the desired fields
            for k, fields in stats_to_keep.items():
                for field in fields:
                    try:
                        video_info[field] = video[k][field]
                    except KeyError:
                        video_info[field] = None  

            all_video_info.append(video_info)

    return all_video_info

json_file = 'video_ids.json' 
playlist_video_data = load_video_ids_from_json(json_file)


#Creating data frames for each playlistID and saving them to my local as csv files
for playlist_id, video_ids in playlist_video_data.items():
    print(f"Processing playlist {playlist_id} with {len(video_ids)} videos.")
    
    video_details = get_video_details(youtube, video_ids)
    df = pd.DataFrame(video_details)
    
    csv_filename = f"{playlist_id}_videos.csv"
    df.to_csv(csv_filename, index=False)
    print(f"Saved {csv_filename} with {df.shape[0]} videos.")

```
