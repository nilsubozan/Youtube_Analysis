```python
import os
import pandas as pd

csv_directory = "csvs"

csv_files = [file for file in os.listdir(csv_directory) if file.endswith('.csv')]

#Initially all columns are uploaded as string to prevent upload error and parsing issues.
csv_dataframes = {}
for file in csv_files:
    file_path = os.path.join(csv_directory, file)
    try:
        csv_dataframes[file] = pd.read_csv(file_path, dtype=str, engine='python')
        print(f"Uploaded: {file}")
    except pd.errors.ParserError as e:
        print(f"Error reading {file}: {e}")

```

```python
top50_wiki = csv_dataframes['top50-youtubers-wiki.csv']
```

```python
top50_wiki
```

```python
#Dropping the first row because it contains NaN
top50_wiki= top50_wiki.drop(0)
```

```python
top50_wiki = top50_wiki.drop(['Subscribers (millions)', 'Link'], axis=1)
```

```python
#Now we are ready for data cleaning.
# '/' will be replaced with ','
#Unnecessary characters and numbers will be removed.
```

```python
#Remove square brackets and numbers from the 'Primary Language' column
top50_wiki['Primary Language'] = top50_wiki['Primary Language'].str.replace(r'[\[\]0-9]', '', regex=True)



```

```python
# Remove numbers and square brackets from the 'Category' column
top50_wiki['Category'] = top50_wiki['Category'].str.replace(r'\[\d+\]', '', regex=True)

top50_wiki['Category'] = top50_wiki['Category'].str.replace('/', ',', regex=False)
```

```python
top50_wiki['Country'] = top50_wiki['Country'].str.replace('/', ',', regex=False)
top50_wiki['Country'] = top50_wiki['Country'].str.replace(r'\[.*?\]', '', regex=True)

```

```python
top50_wiki
```

```python
#Save the cleaned DataFrame to a CSV file
top50_wiki.to_csv('cleaned_top50_wiki.csv', index=False)
```

```python
#Cocomelon Preprocess
cocomelon = csv_dataframes['UUbCmjCuTUZos6Inko4u57UQ_videos.csv']
```

```python
cocomelon
```

```python
#Handling data-time in publishedAt column
cocomelon['publishedAt'] = pd.to_datetime(cocomelon['publishedAt'])
cocomelon['publishedAt'] = cocomelon['publishedAt'].dt.tz_localize(None)
```

```python
#Convert numeric columns
numeric_columns = ['viewCount', 'likeCount', 'favoriteCount', 'commentCount']
cocomelon[numeric_columns] = cocomelon[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
import ast
```

```python
#Function to count the number of tags 
def count_tags(tags_str):
    try:
        tags_list = ast.literal_eval(tags_str)
        if isinstance(tags_list, list):
            return len(tags_list)
        else:
            return 0
    except (ValueError, SyntaxError):
        #If the string is not a valid list, return 0
        return 0
```

```python
cocomelon['tags_count'] = cocomelon['tags'].apply(count_tags)
```

```python
columns = list(cocomelon.columns)

#Finding the index of tags column and positioning in the initial place
tags_index = columns.index('tags')

#Drop the original 'tags' column
cocomelon = cocomelon.drop('tags', axis=1)

#Insert the 'tags_count' column in the original position of 'tags'
columns[tags_index] = 'tags_count'

cocomelon = cocomelon[columns]
```

```python
#Converting duration column from ISO 8601 to minutes
cocomelon['duration'] = cocomelon['duration'].apply(lambda x: isodate.parse_duration(x).total_seconds() / 60)


```

```python
import numpy as np
```

```python
#Some rows gave error because of the invalid strings in duration column.
#This code identifies the rows with invalid duration and replaces them with NaN
#and converts the valid durations to second
invalid_duration_rows = []


def safe_parse_duration(x, idx):
    try:
        return isodate.parse_duration(x).total_seconds() / 60
    except (isodate.ISO8601Error, ValueError, TypeError):
        # If there's an error, log the index of the invalid row
        invalid_duration_rows.append(idx)
        return np.nan  


tseries['duration'] = tseries.apply(lambda row: safe_parse_duration(row['duration'], row.name), axis=1)


print("Rows with invalid durations:", invalid_duration_rows)
```
